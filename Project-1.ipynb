```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop unamed column
df=df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

```python
#gender distribution
plt.figure(figsize=(3,3))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

```python
#from the above chart we have anlysed that the amount of females in the data is morre than the number of  males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean' , "ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Education and Students Score")
plt.show
```

```python
#from the above chart we have concluded that the education of the parents have a good impact
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean' , "ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Marital Status and Student Score")
plt.show
```

```python
#from the above chart we have concluded that the parental status of the parents have no  impact on the students score
```

Distribution Of Ethnic groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
l= ["group A","group B","group C","group D","group E",]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"],groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels=l, autopct = "%1.2f%%")
plt.title("Distribution Of EthnicGroups")
plt.show()
```
